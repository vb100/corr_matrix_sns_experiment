```python
import seaborn as sns
```

<h3>Distracting</h3>

```python
data = sns.load_dataset('diamonds')
data.head(4)
```

```python
# Create a matrix from the data
matrix = data.corr()
```

```python
# Plot the correlation matrix
sns.heatmap(
        matrix,
        center=0,
        fmt='.3f',
        square=True,
        annot=True,
        linewidth=.3
    )
```
